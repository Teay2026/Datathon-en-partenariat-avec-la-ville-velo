```python
import pandas as pd
import folium

geo = pd.read_csv("data/bronze/comptage_manuel/manual_sites_geo.csv")
m = folium.Map(location=[45.77, 4.83], zoom_start=12)

for _, r in geo.iterrows():
    if r.get("status") == "ok":
        folium.Marker(
            location=[r["lat"], r["lon"]],
            popup=f"{r['manual_site_name']}<br>{r.get('display_name','')}"
        ).add_to(m)

m.save("exports/leaflet/manual_sites_preview.html")
print("âœ… map saved: exports/leaflet/manual_sites_preview.html")
```
